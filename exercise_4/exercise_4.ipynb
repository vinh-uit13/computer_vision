```python
import matplotlib.pyplot as plt
from mtcnn.mtcnn import MTCNN
import matplotlib.patches as patches
import cv2
import glob
import numpy as np
from PIL import Image
import os
%matplotlib inline

def extract(b):
  x, y = b[0], b[1]
  width, height = b[2], b[3]
  return x, y, width, height 

if __name__ == '__main__':
   list_img_filename = glob.glob(os.path.join(os.getcwd(), "faces/*.jpg"))
   images = [cv2.imread(file) for file in list_img_filename]
   detector = MTCNN()
   detect_bounds = []
   bbox = []
   for idx,img in enumerate(imgs):
      detect_bound = detector.detect_faces(img)
      detect_bounds.append(detect_bound)  
      img_filename = list_img_filename[idx].split('/')[-1].split('.')[0]
  
      for idx_next, face in enumerate(detect_bound):
          x,y,w,h = extract(face['box'])
          bound = img[y:y+h, x:x+w, :]
          bbox.append(bound)
    
          folder_name = img_filename + 'face_'  + str(idx_next)    
          cv2.imwrite(folder_name + '/'+folder_name+".png", bound)
    
   fig = plt.figure(figsize=(15,90))
   columns = 4
   rows = len(bbox)

   for id, img in enumerate(bbox):
       fig.add_subplot(rows, columns, id+1)
       plt.imshow(img)
       plt.axis('off')

   plt.show()
```
